```python
import numpy as np
print("Total Fraud % ", np.round((data["Class"].value_counts()[1]/data["Class"].value_counts()[0])*100,2))
```

```python
#Splitting dataset into Train and Test set
y = data['Class']
data.drop(columns='Class', inplace=True)
x_train = data[:142400]
x_test = data[142400:227840]
x_val = data[227840:]
y_train = y[:142400]
y_test = y[142400:227840]
y_val = y[227840:]
#Normalization
x_train['Amount'] -= x_train.iloc[:,29:].values.mean()
x_train['Amount'] /= x_train.iloc[:,29:].values.std()

x_test['Amount'] -= x_train.iloc[:,29:].values.mean()
x_test['Amount'] /= x_train.iloc[:,29:].values.std()
```

```python
x_train['Time'] -= x_train.iloc[:,0:1].values.mean()
x_train['Time'] /= x_train.iloc[:,29:].values.std()

x_test['Time'] -= x_train.iloc[:,0:1].values.mean()
x_test['Time'] /= x_train.iloc[:,29:].values.std()
```

```python
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(10, activation = 'relu', input_shape = (30,)))
model.add(layers.Dense(8, activation = 'relu'))
model.add(layers.Dense(6, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history=model.fit(x_train, y_train,validation_data= (x_val, y_val),epochs=100, batch_size=200, verbose=1)
```

```python
#Retraining the model with Regularizer
from keras import models
from keras import layers
from keras import regularizers
model = models.Sequential()
model.add(layers.Dense(10, activation = 'relu', kernel_regularizer= regularizers.l2(0.001), input_shape = (30,)))
model.add(layers.Dense(8, activation = 'relu', kernel_regularizer= regularizers.l2(0.001)))
model.add(layers.Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history=model.fit(x_train, y_train,validation_data= (x_val, y_val),epochs=40, batch_size=200, verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)

prediction = model.predict(x_test)
```
